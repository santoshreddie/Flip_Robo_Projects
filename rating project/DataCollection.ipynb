# Creating script for data collection

Note : just pass the link of the product as the input and set the dataframe column sizes in the data frame section
       i have done only for the flipkart site

```python
#importing necessary libearies
from selenium import webdriver
from bs4 import BeautifulSoup
import requests as rq
import pandas as pd
from time import sleep
import re

#creating lists for storing the data
ratings = []
review = []
comment = []

#creating the input variable for links
url = input("Enter the url : ")
#getting the driver details 
driver = webdriver.Chrome('/home/santosh/Downloads/chromedriver_linux64/chromedriver')
    
#getting the url
driver.get(url)

#scrolling the page down to certain time
driver.execute_script('window.scroll(0, 2500)')

#making driver to pause for a second
sleep(1)

#now cliking on the all reviews hyperlink to navigate to that page
driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]').click()

#after navigating gettig to the page we are getting the url for the current page that we have navigated 
#now we have to create a loop so that we can navigate to all the reviews so that we can scrap the details 
u_rl = driver.current_url

#creating the loop and setting to desired limit
i = 1

while i <= 1000:
    
    #Here there is a bit tricky where we are getting the link and adding a bit of string to the data 
    #so that we can navigate to all the pages till the limit of the loop completed 
    ss = driver.get(str(u_rl)+"&page="+str(i))
    
    #getting the current url
    qq = driver.current_url
    
    #getting all the urls
    r2 = rq.get(qq)
    
    #using beautiful soup we are parsing to parse the string data of the url
    soup = BeautifulSoup(r2.text, 'html.parser')
    
    #getting rating detalis 
    
    
        
        
    
    for ra in soup.find_all('div', {'class' : "_3LWZlK _1BLPMq"}):
            
            aa = ra.get_text() #getting text from the element
        
            ratings.append(aa) #appending to the lists that we have created
        
        
    for ra in soup.find_all('div', {'class' : "_3LWZlK _32lA32 _1BLPMq"}):
            
            aa = ra.get_text() #getting text from the element
        
            ratings.append(aa) #appending to the lists that we have created
            
            
    for ra in soup.find_all('div', {'class' : "_3LWZlK _1rdVr6 _1BLPMq"}):
            
            aa = ra.get_text() #getting text from the element
        
            ratings.append(aa) #appending to the lists that we have created
        
        
        
        
    
    
    sleep(1)        
   
            
    #getting review details         
    for co in soup.find_all('p', {'class' : "_2-N8zT"}):
        
        pc = co.get_text()#getting the text
        
        comment.append(pc)#appending the text
        
    #getting athe comment details    
    for st in soup.find_all('div',  {"class" : "t-ZTKy"}):
        
        fc = st.get_text().replace("\n", " ")#getting the text
        
        review.append(fc)#appending the text
        
        
    
        
    sleep(1) #making the driver to pause for a second
    
    i+=1 #incrementing the loop
    

df=pd.DataFrame({})
df['Rating']=ratings     #creating the data frame from the list data we have obtaines 
df['Review']=review      # a quick note here we have to make the fix the size of the columns
df['Comment'] = comment  #bbecuase we cannot obtail all the details so we have make even sizes                        
df.to_csv("data4.csv")          #so that we might not run into errors

print(df.shape) 
df.head()
```
