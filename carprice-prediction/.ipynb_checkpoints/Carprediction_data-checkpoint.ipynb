```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import time
import re

brand = []
model = []
owners = []
year_of_purchase = []
transmission = []
kilometers = []
fuel_type = []
price = []

driver = webdriver.Chrome("/home/santosh/Downloads/chromedriver_linux64/chromedriver")


driver.get("https://www.cars24.com")

driver.maximize_window()

driver.find_element_by_xpath("/html/body/div[1]/div[5]/div/div/h3/div/img").click()

time.sleep(3)

driver.find_element_by_xpath("/html/body/div[1]/div[1]/div[2]/div/div/div[2]/ul/li[3]/a").click()
 
p = driver.current_url

r1 = requests.get(p)

soup1 = BeautifulSoup(r1.content, 'lxml')

lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
match=False
while(match==False):
    lastCount = lenOfPage
    time.sleep(3)
    lenOfPage = driver.execute_script("window.scrollTo(0, document.body.scrollHeight);var lenOfPage=document.body.scrollHeight;return lenOfPage;")
    if lastCount==lenOfPage:
        match=True
        
time.sleep(5)

nameslist = []
names = driver.find_elements_by_xpath('//h2[@class="_3FpCg"]')
for i in names:
    nameslist.append(i.text)
    
for i in range(len(nameslist)):
    year_of_purchase.append(nameslist[i][:4])
    brand.append(nameslist[i][4:])
    
prices = driver.find_elements_by_xpath('//div[@class="_7udZZ"]')
for i in prices:
    price.append(i.text.replace("₹", "").replace(",", ""))
    
fuel = driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[3]')
for i in fuel:
    fuel_type.append(i.text)
    

km = driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[2]')
for i in km:
    s = i.text
    exp = filter(str.isdigit, s)
    join_str = "".join(exp)
    owners.append(join_str)
    
km = driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[1]')
for i in km:
    kilometers.append(i.text.replace(",", "").replace("km", ""))
    
    
mods = []
models = driver.find_elements_by_xpath('//p[@class="cvakB"]')
for i in models:
    mods.append(i.text)
    
for i in range(len(mods)):
    transmission.append(mods[i].split()[-1])
    
types = []
for i in range(len(models)):
    types.append(brand[i].split()[1:])
    
for i in types:
    listToStr = ' '.join(map(str, i))
    model.append(listToStr)
    
data = pd.DataFrame({})
data["Brand"] = brand
data["Model"] = model
data["Owner"] = owners
data["Kilometer"] = kilometers
data["Year_Of_Purchase"] = year_of_purchase
data["Transmission"] = transmission
data["Fuel_Type"] = fuel_type
data["Price"] = price
    
    
data

data.to_csv("cars24data.csv")
```
