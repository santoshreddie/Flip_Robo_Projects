```python
url = 'https://www.digit.in/slideshows/10-best-gaming-laptops-under-rs-1-lakh-1.html'

#getting responce from the website weather to scrape or no
response = requests.get(url)
print(response)


#getting the information of driver for accesig the driver for automation
driver = webdriver.Chrome('/home/santosh/Downloads/chromedriver_linux64/chromedriver')
#naviagting to the url
driver.get(url)

time.sleep(4)#making browser wait for sometime

commandata = []
commanclass = driver.find_elements_by_xpath('//div[@class="col-sm-12"]')
#getting the details of top 10 laptops
for i in commanclass:
    commandata.append(i.text)
    
#the whole data contained mixed values so ay first i will remove the 

new_data = list(filter(None, commandata))[1:]#to remove the empty values in the data and 1 st element 
                                             #which is just introduction
    
#creating empty list to store data

names = []
prices = []
descriptions = []
displays = []
CPU = []
GPU = []
RAM = []
storage = []

#now appending data into the empty lists

for i in range(len(new_data)):
    name = new_data[i].split("\n")[0]
    names.append(name)
    
    price = new_data[i].split("\n")[1]
    newprice = re.findall('\d+\,\d+', price)
    prices.append(newprice)
    
    descri = new_data[i].split("\n")[2]
    descriptions.append(descri)
    
    display = new_data[i].split("\n")[3]
    newdisplay = re.split('\s+', display)[1]
    displays.append(newdisplay)
    
    Cpu = new_data[i].split("\n")[4]
    newcpu = re.split('\:+', Cpu)[1]
    CPU.append(newcpu)
    
    Gpu = new_data[i].split("\n")[5]
    newgpu = re.split('\:+',Gpu)[1]
    GPU.append(newgpu)
    
    ram = new_data[i].split("\n")[6]
    newram = re.split('\:+', ram)[1]
    RAM.append(newram)
    
    store = new_data[i].split("\n")[7]
    newstore = re.split('\:+','Storage: 1TB HDD')[1]
    storage.append(newstore)
    
    #creating data with strored data
    
laptops = pd.DataFrame({})

laptops["Name"] = names
laptops["Price"] = prices
laptops["Descriptions"] = descriptions
laptops["Display"] = displays
laptops["CPU"] = CPU
laptops["GPU"] = GPU
laptops["RAM"] = RAM
laptops["Storage"] = storage

#saving data to csvformat

laptops.to_csv("laptop.csv")

laptops
    
```

```python
#importing necessary libraries
from selenium import webdriver
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium.common.exceptions import *

#setting the url
url = "https://trak.in/india-startup-funding-investment-2015/"

response = requests.get(url)
print(response)

#getting the information of driver for accesig the driver for automation
driver = webdriver.Chrome('/home/santosh/Downloads/chromedriver_linux64/chromedriver')
#naviagting to the url
driver.get(url)

#creating empty lists for storing the data

header = []
date = []
startup = []
industry = []
subvertical = []
city = []
investor = []
investment = []
amount = []

#gettin all html tag data for processing actually we have to get thee tables so getting 3 tagof three tables
#tables of july-20 - aug-20
soup = BeautifulSoup(response.content, 'html.parser')
tags = soup.find('table', class_="tablepress tablepress-id-50")
headers = tags.find_all('th')
for i in headers:
    header.append(i.text)
datum1 = []   
data1 = tags.find_all('td')
for i in data1:
    datum1.append(i.text.strip())
    
#table 2
    
tags2 = soup.find('table', class_="tablepress tablepress-id-49")
datum2 = []   
data2 = tags2.find_all('td')
for i in data2:
    datum2.append(i.text.strip())
    
#table 3
    
tags3 = soup.find('table', class_="tablepress tablepress-id-48")
datum3 = []   
data3 = tags3.find_all('td')
for i in data3:
    datum3.append(i.text.strip())
    
#mixing all the tag data into a single list so that we can process easily
    
new_datum = datum1+datum2+datum3

#indexing the liso to get extract required elements 

date = new_datum[1::9]
startup = new_datum[2::9]
industry = new_datum[3::9]
subvertical = new_datum[4::9]
city = new_datum[5::9]
investor = new_datum[6::9]
investment = new_datum[7::9]
amount = new_datum[8::9]

#ceating the dataframe to store the data

import pandas as pd
cycle = pd.DataFrame({})
cycle['Date (dd/mm/yyyy)'] = date
cycle['Startup Name'] = startup
cycle['Industry / Vertical'] = industry
cycle['Sub-Vertical'] = subvertical
cycle['City / Location'] = city
cycle["Investors' Name"] = investor
cycle['Investment Type'] = investment
cycle['Amount (In USD)'] = amount
    
#saving data to csv
cycle.to_csv('cycle.csv')

#showing the detals
cycle

#the list will be in the decending order from aug to july
```
