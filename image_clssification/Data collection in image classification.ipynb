```python
from selenium import webdriver
import requests
import time


driver=webdriver.Chrome("/home/santosh/Downloads/chromedriver_linux64/chromedriver") 
time.sleep(3)

url = "https://www.flipkart.com/"
driver.get(url)

time.sleep(2)
#locating the search bar

driver.find_element_by_xpath("/html/body/div[2]/div/div/button").click()

time.sleep(2)

search_bar=driver.find_element_by_class_name("_3704LK")
search_bar.send_keys('trousers for men')

time.sleep(2)
#locating the button and clicking it toh search for sunglasses
button=driver.find_element_by_class_name('L0Z3Pu').click()


```

```python
urls = []
start = 0
end = 60

for page in range(start,end):
    images = driver.find_elements_by_xpath('//img[@class="_2r_T1I"]')
    for image in images:
        urls.append(image.get_attribute('src'))
        
for i in range(len(urls)):
    if i >= 2399:
        break
    print("Downloading {0} of {1} images" .format(i, 2399))
    response= requests.get(urls[i])
    file = open(r"/home/santosh/data/trousers/trouser"+str(i)+".jpg", "wb")
    file.write(response.content)    

```
